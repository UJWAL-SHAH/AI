```
#Reading the libraries to be used
import pandas as pd
import matplotlib.pylab as plt 
import seaborn as sns
```

```
class DataRead1:
  def __init__(self,path):
     self.data1=pd.read_csv(path)  #PART 1 

  def rowselect(self,x,y):  #PART 2
    self.x=x
    self.y=y
    return (self.data1.iloc[[self.x,self.y]])
    
class DataRead2(DataRead1):
  def correlation(self):
    fulldata=self.data1.corr()
    sns.heatmap(fulldata)
    plt.show()
    
  def correlation2(self,col1,col2):
    self.col1=col1
    self.col2=col2
    sns.regplot(x=self.data1.iloc[:,self.col1] ,y=self.data1.iloc[:,self.col2])

  def rowselect(self,x,y,z):  #PART 4: verwriting the rowselect function of Dataread1 
    self.x=x
    self.y=y
    self.z=z
    return (self.data1.iloc[[self.x,self.y,self.z]])
```
